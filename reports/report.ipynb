## Kategorisierung der Metriken:
Bei diesen Daten sind Metriken aufgeführt, die in unterschiedliche Kategorien aufgeteilt werden können. Es gibt die **beschreibenden Metriken** eines Spielers, wie beispielsweise den Namen, das Alter oder das Team, in dem der Spieler in der betrachteten Saison gespielt hat. Eine weitere Kategorie sind die **traditionellen Metriken**. Diese beinhalten Werte wie GP (Games Played), PPG (Points Per Game), APG (Assists Per Game), FTA (Free Throw Attempts), 2PA (Two-Point Attempts) usw. Diese Metriken werden durch Aktionen während eines Spiels ermittelt, die von der NBA aufgezeichnet und bereitgestellt werden. Die dritte Kategorie sind die **fortgeschrittenen Statistiken**, wie beispielsweise ORtg (Offensive Rating), DRtg (Defensive Rating) und VI (Versatility Index). Für diese werden spezifische Berechnungsformeln genutzt, um generelle Statistiken zu erstellen, die das offensive Potenzial (ORtg) bzw. das defensive Potenzial (DRtg) eines Spielers abbilden.

## Fokus des Projektes:
Der Fokus dieses Projektes liegt darin, herauszufinden, welche Metriken den größten Einfluss auf die fortgeschrittenen Leistungsmetriken (VI, ORtg und DRtg) haben. Ziel ist es, zu identifizieren, welche Metriken bereits mit den fortgeschrittenen Leistungsmetriken in diesem Datensatz abgedeckt sind und welche durch die Aufnahme weiterer Metriken ergänzt werden sollten.

## Forschungsfrage:

# Ausreißer-Analyse und Bereinigung

Um potenzielle Ausreißer in den Datensätzen zu identifizieren, wurde die Methode der **3-fachen Standardabweichung** angewendet.

### 1. Berechnung der Ausreißergrenzen
- **Untergrenze** = Mittelwert - 3 * Standardabweichung
- **Obergrenze** = Mittelwert + 3 * Standardabweichung

### 2. Bewertung der gefundenen Ausreißer:

- **VI, ORtg, DRtg**: In den Response-Variablen wurden 0-Werte gefunden. Diese Werte stellen keine sinnvollen Messwerte dar und müssen daher aus den Datensätzen entfernt werden, da sie die Ergebnisse verfälschen würden.
  
- **USG%**: Ein weiteres auffälliges Merkmal ist, dass einige Werte für **USG%** (Nutzungsrate eines Spielers) zwischen 33% und 40% liegen. Diese Werte erscheinen jedoch realistisch für wichtige Spieler eines Teams. Ein Spieler hat jedoch einen **USG% von 64%**, obwohl dieser Spieler nur **ein einziges Spiel** gespielt hat (etwas mehr als 3 Minuten Spielzeit). Dieser Wert ist daher ein klarer Ausreißer und wird als nicht realistisch erachtet. Um solche Ausreißer zu vermeiden, wird eine **Minimalgrenze von 10 Spielen** eingeführt, die ein Spieler bestritten haben muss, damit seine Werte als gültig betrachtet werden.

### 3. Weitere Ausreißer:
- Die restlichen Werte, die möglicherweise als Ausreißer erscheinen, sind entweder:
    - **Keine Ausreißer**, sondern die besten Spieler der NBA (mit realistischen hohen Werten) oder
    - Spieler mit **geringen spezifischen Metriken**, die auf den ersten Blick untypisch erscheinen, jedoch keine realen Ausreißer darstellen.
  
### 4. Anpassung der Spalten:
- Die Spalte "RANK" wurde entfernt und durch eine eindeutige **ID** ersetzt, um jedem Eintrag eine valide Identifikation zuzuweisen. Dies sorgt für eine klare Zuordnung der Datenpunkte.

Diese Schritte stellen sicher, dass alle unplausiblen Ausreißer entfernt werden und die verbleibenden Daten realistisch und aussagekräftig sind, um präzise Analysen und Modellierungen durchzuführen.

```python
# Outlier Berechnung mit 3*std
std_outliers = {}
outlierList = []
outlier_df = pd.DataFrame(columns=['Column', 'Mean', 'Std Dev', 'Lower Bound', 'Upper Bound', 'Outlier Value'])
for col in df.select_dtypes(include=[np.number]).columns:
    mean = df[col].mean()
    std = df[col].std()
    lower_bound = mean - 3 * std 
    upper_bound = mean + 3 * std

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    std_outliers[col] = outliers

    if len(outliers) > 0:
        outlierList.append(col)
        new_rows = []
        new_rows.append({
            'Column': col,
            'Mean': round(mean, 2),
            'Std Dev': round(std, 2),
            'Lower Bound': round(lower_bound, 2),
            'Upper Bound': round(upper_bound, 2),
            'Outlier Value': outliers.tolist()
        })
        outlier_df_new = pd.DataFrame(new_rows)
        outlier_df = pd.concat([outlier_df, outlier_df_new], ignore_index=True)

print(outlier_df)
```

# Methodology:
## 1. Verteilungsanalyse der Zielvariablen:
- Die Verteilung der VI-Werte zeigt eine leicht linkssteile (rechtsschiefe) Verteilung. Dies bedeutet, dass die Werte eine Asymmetrie aufweisen, was die Anwendung bestimmter Methoden wie Lasso oder LinearRegression beeinflussen könnte. Bei rechtsschiefen Verteilungen könnten auch Transformationen hilfreich sein, um die Verteilung zu stabilisieren.
- ORtg und DRtg hingegen haben eine Verteilung, die sehr nahe an der Normalverteilung liegt. Normalverteilte Daten sind besonders geeignet für lineare Modelle, da diese Modelle annehmen, dass Fehler normalverteilt sind, was bei diesen beiden Variablen zutrifft.

## 2. Modellwahl:
- Die beiden Hauptmethoden, die für diesen Datensatz in Betracht gezogen werden, sind LinearRegression und Lasso Regression. Beide Modelle sind lineare Modelle, jedoch hat Lasso einen zusätzlichen Regularisierungsterm, der hilft, Überanpassung (Overfitting) zu vermeiden, indem er die Koeffizienten bestimmter Variablen auf null setzt.
- Da die Prädiktorvariablen in den Daten stark unterschiedliche Wertebereiche haben (z. B. Werte zwischen 0 und 1 wie 3P% und Werte zwischen 0 und 1500 wie FTA), wurde zusätzlich der StandardScaler verwendet, um alle Variablen auf denselben Skalenbereich zu bringen. Dies ist besonders wichtig für die Lasso-Regression, da die Regularisierung durch den Unterschied in der Skala der Variablen verzerrt werden könnte.

## 3. Modellierung und Bewertung:
- Modellerstellung: Für jede Zielvariable (Response Variable) wurde eine Liste mit den 10 stärksten Prädiktoren (auf Basis der Korrelation) erstellt. Diese Liste enthält die Variablen, die den höchsten Zusammenhang mit der jeweiligen Zielvariable aufweisen.
- Kombinationen testen: Alle Kombinationen dieser 10 Prädiktorvariablen wurden genutzt, um lineare Regressionsmodelle zu erstellen. Diese Modelle wurden dann auf den Testdaten evaluiert.
- Modellbewertung: Die Modelle wurden anhand der folgenden Kennzahlen bewertet:
    - R² (Bestimmtheitsmaß): Gibt an, wie gut das Modell die Varianz der Zielvariable erklärt.
    - MSE (Mean Squared Error): Misst den durchschnittlichen Fehlerquadratswert, der angibt, wie weit die Vorhersagen vom tatsächlichen Wert entfernt sind.
- Ergebnisse speichern: Die Ergebnisse der Modelle, einschließlich der genutzten Prädiktorvariablen, der Zielvariable, der Anzahl der Prädiktoren und des Modells (z. B. "Response Variable" + "Modell Linear"), wurden in einem DataFrame gespeichert.
## 4. Skalierte Modelle
- In einer weiteren Phase wurde die Skalierung der Daten durch den StandardScaler ergänzt. Für jedes Modell wurden sowohl unskalierte als auch skalierte Trainings- und Testdaten verwendet, um zu überprüfen, ob die Skalierung eine signifikante Verbesserung der Modellleistung bringt.
- Diese Methodik wurde auf vier verschiedene Modelle angewendet (einschließlich der Linear- und Lasso-Regression), wobei die StandardScaler-Methode in den zweiten Varianten der linearen und Lasso-Regression integriert wurde.

## 5. Ergebnisspeicherung
- Nach der Modellbewertung wurden alle Ergebnisse in einem DataFrame gespeichert. Dieser DataFrame wurde für jede Zielvariable sortiert, sodass die besten Modelle leicht identifiziert werden konnten. ## Generelles zu den Daten:
Die Daten sind im Großen und Ganzen gut für lineare Regression geeignet. Es gab jedoch einige Probleme, die erfolgreich behoben werden konnten. Dazu zählen:

1. **Einige 0-Werte in den Predictor-Variablen**
   - Diese wurden durch Filtern auf eine minimale Anzahl gespielter Spiele behandelt.
   - 0-Werte können jedoch realistisch sein, z. B. bei "FTA", wenn ein Spieler nicht gefoult wurde. Deshalb wurde dies im Kontext der Spielerrolle und Spielzeit berücksichtigt.

2. **Nicht realistische Ausreißer**
   - Diese wurden durch Filtern auf eine minimale Anzahl an Spielen behandelt, die ein Spieler absolviert haben muss.

3. **Einige 0-Werte in den Response-Variablen**
   - Spieler mit einem 0-Wert in einer Response-Variablen wurden nach Anwendung der Filter auf die Anzahl gespielter Spiele entfernt.

4. **Fehlende Werte in der Spalte `RANK`**

5. **Nicht skalierte Daten**
